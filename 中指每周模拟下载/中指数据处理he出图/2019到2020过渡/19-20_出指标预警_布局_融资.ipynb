````python
# 读取数据

deal = pd.read_excel('../数据整理/中指综合后的2016-2020数据.xlsx', sheet_name='土地成交明细')
# deal = deal[deal['周'] != currentWeek+1]
liupai = pd.read_excel('../数据整理/中指综合后的2016-2020数据.xlsx', sheet_name='土地流拍')
# liupai = liupai[liupai['周'] != currentWeek+1]

deal.replace('--', 0, inplace=True)
liupai.replace('--', 0, inplace=True)


deal.loc[deal[deal['起始价(万元)']==0].index, '起始价(万元)'] = deal[deal['起始价(万元)']==0]['成交价(万元)']
deal.loc[deal[deal['成交价(万元)']==0].index, '成交价(万元)'] = deal[deal['成交价(万元)']==0]['起始价(万元)']

deal.fillna(0, inplace=True)
liupai.fillna(0, inplace=True)
````

### 图1 101城溢价率结构表__按周城市出

**计算 101城, 12线, 34线城市 溢价率 **

````python
# TODO
# 本周每个城市溢价率分类详情
d = deal_liupai[deal_liupai['周']==currentWeek]
a = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市', fill_value=0)
a.head(2)
````

````python
# 计算每个城市溢价率   去除 流拍选项
dd = d[d['溢价率分类'] != '流拍']
d1 = dd.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市', fill_value=0)
d1['溢价率'] = (d1['成交价(万元)'] / d1['起始价(万元)'] - 1) *100
d1 = d1[['溢价率']]

# 计算101城市溢价率
d1.loc['101城'] = (dd['成交价(万元)'].sum() / dd['起始价(万元)'].sum() - 1) *100

# 计算12,和 34 线城市溢价率
d1234 = dd.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市能级', fill_value=0)
d1234['溢价率'] = (d1234['成交价(万元)'] / d1234['起始价(万元)'] - 1) *100
d1234 = d1234[['溢价率']]
d1234


## 新加1.1  计算 城市类型的溢价率
dType = dd.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市分类', fill_value=0)
dType['溢价率'] = (dType['成交价(万元)'] / dType['起始价(万元)'] - 1) *100
dType = dType[['溢价率']]
dType

# 合并 全国和12,34线城市
d_all_city = pd.concat((d1, d1234, dType))
d_all_city.replace([np.inf,-np.inf],0, inplace=True)
d_all_city.tail()
````

**计算101, 12, 34 城市 成交率**

````python
## 新加1.1 城市类型
dType_city = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市分类', fill_value=0)

# 12, 34 线
a12 = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市能级', fill_value=0)

# 101城
a101 = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count', fill_value=0)
a101.index = ['101城']
display(dType_city,a101, a12)


# 全国和101,3412城市综合列表
a101_12_34 = pd.concat((a,a12,a101, dType_city))


# 将全国溢价率 合并到全国各个城市中
all_city = pd.concat((a101_12_34,d_all_city), axis=1, sort=True)
all_city.head()
````

````python
# 101 最终状态, 预留和后边底稿合并
c = all_city.copy()
if '流拍' not in c.columns:
    c['流拍'] = 0
if '高溢价' not in c.columns:
    c['高溢价'] = 0
c.eval(' 总计 = 底价+溢价+高溢价', inplace=True)
c.eval(' 总计流拍 = 底价+溢价+流拍+高溢价', inplace=True)

c['底价成交率'] = (c['底价'] / c['总计']).apply(lambda x: format(x, '.4f')) 
c['流拍率'] = (c['流拍'] / c['总计流拍']).apply(lambda x: format(x, '.4f')) 
# c['溢价'] = (c['溢价'] / c['总计']).apply(lambda x: format(x, '.4f')) 
c['高溢价成交率'] = (c['高溢价'] / c['总计']).apply(lambda x: format(x, '.4f')) 
c = c[['底价成交率', '高溢价成交率', '流拍率','溢价率']]
c.head()
````
