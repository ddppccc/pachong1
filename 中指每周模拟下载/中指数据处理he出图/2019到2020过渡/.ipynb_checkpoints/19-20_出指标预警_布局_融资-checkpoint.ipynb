````python
import pandas as pd
from prettytable import PrettyTable
from PIL import Image, ImageDraw, ImageFont

def create_table_img(data,img_name, **kwargs):
    '''
        img_name 图片名称 'D:/project/pythonwork/12306/t.png' 或 t.png
        data 表格内容，首行为表头部
        table_title 表格标题
        line_height 底部描述行高
        font 默认字体路径
        default_font_size 默认字体大小
        default_background_color 图片背景底色
        table_top_heght 设置表格顶部留白高度
        table_botton_heght 设置表格顶部留白高度
        describe 底部描述文字
    '''
    space = 0 ## 表格边距
    # 生成图片-------------------------------
    ### 底部描述行高
    if 'line_height' not in kwargs:  
         # line_height = 1
         line_height = 0
    else:
        line_height = kwargs['line_height']

    ### 默认字体
    if 'font' not in kwargs:  
         kwargs['font'] = None

    ### 默认字体大小
    if 'default_font_size' not in kwargs:
         kwargs['default_font_size'] = 12

    ### 默认表标题字体大小
    if 'table_title_font_size' not in kwargs:
#          kwargs['table_title_font_size'] = 22
         kwargs['table_title_font_size'] = 0


    ### 图片背景底色
    if 'default_background_color' not in kwargs:
         kwargs['default_background_color'] = (255,255,255,255)

    ### 设置表格顶部留白高度
    if 'table_top_heght' not in kwargs:
         kwargs['table_top_heght'] = kwargs['table_title_font_size'] + space + int(kwargs['table_title_font_size']/2)

    ## 底部描述文字
    if 'describe' in kwargs:
        describe_len = len(kwargs['describe'])
    else:
        describe_len = 0

    ### 设置表格底部留白高度
    if 'table_botton_heght' not in kwargs:
         kwargs['table_botton_heght'] = describe_len * kwargs['default_font_size'] + space
    
    ### 图片后缀
    if 'img_type' not in kwargs:
         kwargs['img_type'] = 'PNG'

    ### 默认字体及字体大小
    font = ImageFont.truetype(kwargs['font'], kwargs['default_font_size'], encoding='utf-8')
    font2 = ImageFont.truetype(kwargs['font'], kwargs['table_title_font_size'], encoding='utf-8')
    ## Image模块创建一个图片对象
    im = Image.new('RGB',(1, 1),kwargs['default_background_color'])
    ## ImageDraw向图片中进行操作，写入文字或者插入线条都可以
    draw = ImageDraw.Draw(im)
       
    # 创建表格---------------------------------
    tab = PrettyTable(border=True, header=True, header_style = 'title')
    ## 第一行设置为表头
    tab.field_names = data.pop(0)
    for row in data:
        tab.add_row(row)
    tab_info = str(tab)
    ## 根据插入图片中的文字内容和字体信息，来确定图片的最终大小
    img_size = draw.multiline_textsize(tab_info, font=font)
#     img_width = img_size[0]+space*2
#     table_height = img_size[1]+space*2
    img_width = img_size[0]
    table_height = img_size[1]
    img_height = table_height + kwargs['table_botton_heght'] + kwargs['table_top_heght']
    im_new = im.resize((img_width, img_height))
#     im_new = im.resize((405, 340))
    del draw
    del im
    draw = ImageDraw.Draw(im_new, 'RGB')
    draw.multiline_text((space,kwargs['table_top_heght']), tab_info + '\n\n', fill=(0,0,0), font=font)

    ### 表标题--------------------------
    if 'table_title' in kwargs:
        title_left_padding = (img_width - len(table_title) * kwargs['table_title_font_size']) / 2
        draw.multiline_text((title_left_padding, space), table_title, fill=(16,0,0), font=font2, align = 'center')


    y = table_height + space/2

    ### 描述内容-----------------------------------
    if 'describe' in kwargs:
        y = y + kwargs['default_font_size']
        frist_row = kwargs['describe'].pop(0)
        draw.text((space,y), frist_row, fill=(255,0,0), font=font)
        for describe_row in kwargs['describe']:
            y = y + kwargs['default_font_size'] + line_height
            draw.text((space,y), describe_row, fill=(0,0,0), font=font)
    del draw
    ### 保存为图片
    im_new.save(img_name,  kwargs['img_type'], quality=200)
    return True

````

### 图1 101城溢价率结构表__按周城市出

**计算 101城, 12线, 34线城市 溢价率 **

````python
# TODO
# 本周每个城市溢价率分类详情
d = deal_liupai[deal_liupai['周']==currentWeek]
a = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市', fill_value=0)
a.head(2)
````

````python
# 计算每个城市溢价率   去除 流拍选项
dd = d[d['溢价率分类'] != '流拍']
d1 = dd.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市', fill_value=0)
d1['溢价率'] = (d1['成交价(万元)'] / d1['起始价(万元)'] - 1) *100
d1 = d1[['溢价率']]

# 计算101城市溢价率
d1.loc['101城'] = (dd['成交价(万元)'].sum() / dd['起始价(万元)'].sum() - 1) *100

# 计算12,和 34 线城市溢价率
d1234 = dd.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市能级', fill_value=0)
d1234['溢价率'] = (d1234['成交价(万元)'] / d1234['起始价(万元)'] - 1) *100
d1234 = d1234[['溢价率']]
d1234


## 新加1.1  计算 城市类型的溢价率
dType = dd.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市分类', fill_value=0)
dType['溢价率'] = (dType['成交价(万元)'] / dType['起始价(万元)'] - 1) *100
dType = dType[['溢价率']]
dType

# 合并 全国和12,34线城市
d_all_city = pd.concat((d1, d1234, dType))
d_all_city.replace([np.inf,-np.inf],0, inplace=True)
d_all_city.tail()
````

**计算101, 12, 34 城市 成交率**

````python
## 新加1.1 城市类型
dType_city = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市分类', fill_value=0)

# 12, 34 线
a12 = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市能级', fill_value=0)

# 101城
a101 = d.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count', fill_value=0)
a101.index = ['101城']
display(dType_city,a101, a12)


# 全国和101,3412城市综合列表
a101_12_34 = pd.concat((a,a12,a101, dType_city))


# 将全国溢价率 合并到全国各个城市中
all_city = pd.concat((a101_12_34,d_all_city), axis=1, sort=True)
all_city.head()
````

````python
# 101 最终状态, 预留和后边底稿合并
c = all_city.copy()
if '流拍' not in c.columns:
    c['流拍'] = 0
if '高溢价' not in c.columns:
    c['高溢价'] = 0
c.eval(' 总计 = 底价+溢价+高溢价', inplace=True)
c.eval(' 总计流拍 = 底价+溢价+流拍+高溢价', inplace=True)

c['底价成交率'] = (c['底价'] / c['总计']).apply(lambda x: format(x, '.4f')) 
c['流拍率'] = (c['流拍'] / c['总计流拍']).apply(lambda x: format(x, '.4f')) 
# c['溢价'] = (c['溢价'] / c['总计']).apply(lambda x: format(x, '.4f')) 
c['高溢价成交率'] = (c['高溢价'] / c['总计']).apply(lambda x: format(x, '.4f')) 
c = c[['底价成交率', '高溢价成交率', '流拍率','溢价率']]
c.head()
````

````python
# TODO
# 本周每个城市溢价率分类详情
d_yue = deal_liupai[deal_liupai['月']==lastMonth]
a_yue = d_yue.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市', fill_value=0)
a_yue.head()



# 计算每个城市溢价率   去除 流拍选项
dd_yue = d_yue[d_yue['溢价率分类'] !='流拍']
d1_yue = dd_yue.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市', fill_value=0)
d1_yue['溢价率'] = (d1_yue['成交价(万元)'] / d1_yue['起始价(万元)'] - 1) *100
d1_yue = d1_yue[['溢价率']]

d1_yue.head()



# 计算101城市溢价率
d1_yue.loc['101城'] = (dd_yue['成交价(万元)'].sum() / dd_yue['起始价(万元)'].sum() - 1) *100

# 计算12,和 34 线城市溢价率
d1234_yue = dd_yue.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市能级', fill_value=0)
d1234_yue['溢价率'] = (d1234_yue['成交价(万元)'] / d1234_yue['起始价(万元)'] - 1) *100
d1234_yue = d1234_yue[['溢价率']]
d1234_yue.head()


## 新加1.1  计算 城市类型的溢价率
dType_yue = dd_yue.pivot_table(values=['起始价(万元)','成交价(万元)'], aggfunc='sum',index='城市分类', fill_value=0)
dType_yue['溢价率'] = (dType_yue['成交价(万元)'] / dType_yue['起始价(万元)'] - 1) *100
dType_yue = dType_yue[['溢价率']]
dType_yue



# 合并 全国和12,34线城市
d_all_city_yue = pd.concat((d1_yue, d1234_yue, dType_yue))
d_all_city_yue.replace([np.inf,-np.inf],0, inplace=True)
d_all_city_yue.head()



## 新加1.1 城市类型
dType_city_yue = d_yue.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市分类', fill_value=0)


# 12, 34 线
a12_yue = d_yue.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count',index='城市能级', fill_value=0)

# 101线
a101_yue = d_yue.pivot_table(values='地块名称', columns='溢价率分类', aggfunc='count', fill_value=0)
a101_yue.index = ['101城']
display(dType_city_yue, a101_yue, a12_yue)

# 全国和101,3412城市综合列表
a101_12_34_yue = pd.concat((a_yue, a12_yue, a101_yue, dType_city_yue))

# 合并溢价分类, 和 溢价率数据
all_city_yue = pd.concat((a101_12_34_yue, d_all_city_yue), axis=1, sort=True)
all_city_yue.head()
````
